```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Lisez le fichier CSV en spécifiant la colonne qui contient les noms de pays
df = pd.read_csv("world_population.csv")

# Excluez les colonnes non numériques (par exemple, les noms de pays)
dff = df.select_dtypes(include=[int, float])

 

# Calculez la matrice de corrélation
correlation_matrix = dff.corr()

# Configurez la taille de la figure
plt.figure(figsize=(12, 8))

# Affichez la heatmap de corrélation avec des annotations
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap de Corrélation')
plt.show()

 

```

```python
import pandas as pd

# Lire le DataFrame depuis un fichier CSV
df = pd.read_csv("world_population.csv")
# Renommez les colonnes avec des noms numériques
df = df.rename(columns={'2022 Population': 'Population_2022',
                        '2020 Population': 'Population_2020',
                        '2015 Population': 'Population_2015',
                        '2010 Population': 'Population_2010',
                        '2000 Population': 'Population_2000',
                        '1990 Population': 'Population_1990',
                        '1980 Population': 'Population_1980',
                        '1970 Population': 'Population_1970',
                        'Area (km²)': 'Area_km2',
                        'Density (per km²)': 'Density_per_km2',
                        'Growth Rate': 'Growth_Rate'})
df.head()
 
# Supprimer les colonnes non nécessaires
#df = df.drop(columns=['Country Code', 'Country Name'])

# Convertir les colonnes numériques en types de données numériques
numeric_columns = ['Population_2022', 'Population_2020', 'Population_2015', 'Population_2010', 'Population_2000', 'Population_1990', 'Population_1980', 'Population_1970', 'Area_km2', 'Density_per_km2', 'Growth_Rate']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Remplacer les valeurs manquantes par des zéros
df = df.fillna(0)

# Réexécuter l'opération de calcul de la moyenne
df.groupby('Continent').mean()

 
```
